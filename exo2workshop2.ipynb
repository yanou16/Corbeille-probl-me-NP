Is Knights ∈ NP ??

We show that the Knights of the Round Table problem is in NP.

A certificate for this problem is an ordering of the knights around the table.
To verify the certificate, the algorithm must:

Check that each knight appears exactly once in the ordering.

For each consecutive pair of knights (including the last and the first), check that they are not enemies.

Assuming that checking if two knights are enemies can be done in constant time (using an adjacency matrix), the verification requires O(n) checks for adjacency and O(n) for uniqueness, which is polynomial in n.

Therefore, Knights ∈ NP.

On peut vérifier une solution (un ordre des chevaliers) rapidement → donc c’est bien dans NP.

```python
import networkx as nx
import matplotlib.pyplot as plt

# --- Exemple 1 : 4 chevaliers avec quelques ennemis ---
G1 = nx.Graph()
G1.add_nodes_from(["A", "B", "C", "D"])
# Arêtes = "amis" (pas ennemis)
G1.add_edges_from([("A", "B"), ("B", "C"), ("C", "D"), ("D", "A")])

plt.figure(figsize=(4,4))
nx.draw_circular(G1, with_labels=True, node_color='lightblue', node_size=1000, font_weight='bold')
plt.title("Graphe 1 : 4 chevaliers formant un cycle possible")
plt.show()


# --- Exemple 2 : ajout d’un ennemi (donc arête manquante) ---
G2 = nx.Graph()
G2.add_nodes_from(["A", "B", "C", "D"])
G2.add_edges_from([("A", "B"), ("B", "C"), ("C", "D")])  # Pas de (D, A)

plt.figure(figsize=(4,4))
nx.draw_circular(G2, with_labels=True, node_color='lightcoral', node_size=1000, font_weight='bold')
plt.title("Graphe 2 : un cycle impossible (A et D sont ennemis)")
plt.show()


# --- Exemple 3 : graphe complet (tout le monde ami avec tout le monde) ---
G3 = nx.complete_graph(5)
labels = {i: f"K{i+1}" for i in range(5)}

plt.figure(figsize=(4,4))
nx.draw_circular(G3, labels=labels, with_labels=True, node_color='lightgreen', node_size=1000, font_weight='bold')
plt.title("Graphe 3 : chaque chevalier peut s’asseoir à côté de n’importe qui")
plt.show()
```
